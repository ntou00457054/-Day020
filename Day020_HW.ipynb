```python
from bs4 import BeautifulSoup #BeautifulSoup 取出html結構

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
data = {}

for tr in table.find_all('tr')[1:]:
    for i ,td in enumerate(tr.find_all('td')):
        if len(tr.find_all('td')) == 5:
            
            if i == 0:
                test_item = td.get_text()
                data.setdefault(test_item,{})
            if i==2:
                date = td.get_text()
            if i==3:
                value = td.get_text()
                data[test_item][date] = value
                
        if len(tr.find_all('td')) == 3:
            if i==0:
                date = td.get_text()
            if i==1:
                value = td.get_text()
                data[test_item][date] = value
print(data)
                
        
```
